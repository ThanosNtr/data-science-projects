```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_exoplanets_2021.csv')
```

```python
df.isnull().sum()
```

```python
planet_col = ['Planet Name', 'Num Stars','Num Planets', 'Discovery Method', 'Discovery Year', 'Orbital Period Days', 'Discovery Facility']
```

```python
df1=df[planet_col]
```

```python
df1.head()
```

```python
Star_count=df1.sort_values(['Num Stars'], ascending=False)
Star_count=pd.DataFrame(Star_count)
Star_count=Star_count[0:10]
Star_count
```

```python
#COUNT HOW MANY STARS PLANET HAVE
Star_count.plot.bar(x='Planet Name', y='Num Stars')
```

```python
# HOW WAS PLANET DISCOVERED
df1 = df["Discovery Method"].value_counts().reset_index()
df1.columns=['Method','Count']
df1
```

```python

sns.catplot(x='Count', y='Method', kind='bar', data=df1 , height=6 , aspect=3)
plt.grid(True)
```

```python
#Planets Discovered every Year
df1 = df["Discovery Year"].value_counts().reset_index()
df1.columns = ["Year", "Count"]
df1
```

```python
sns.catplot(y='Count', x='Year', kind='bar', data=df1, height=7, aspect=3)
plt.grid(True)
```

```python
df['Discovery Facility'].value_counts()
```

```python
#Top 10 Facilities
df1=df['Discovery Facility'].value_counts()[0:10].reset_index()
df1.columns = ['Facility','Count']
df1
```

```python
plt.pie(df1['Count'],labels = df1['Facility'],autopct='%1.1f%%',radius=3,shadow=True)
plt.show()
```
